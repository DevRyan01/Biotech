Validação e Inferência com YOLO11-OBB

Este notebook demonstra como:
- Validar o modelo treinado
- Realizar inferência (predição) em uma imagem

Cada célula está documentada em detalhes para compreensão completa.

## 1. Importações

````python
from ultralytics import YOLO
from PIL import Image
````

## 2. Carregamento do Modelo Base

- Carrega o modelo base YOLO11m-OBB (pré-treinado)
- A versão "m" é intermediária (balanceia desempenho e velocidade)
- O sufixo "obb" indica suporte a bounding boxes orientadas

````python
model = YOLO('best.pt')
````

## 3. Validação do Modelo

Carrega o modelo treinado e calcula métricas de desempenho.

````python
results = model.val(imgsz=640, batch=32, conf=0.001, iou=0.7, save_json=False, split='test')
print(results)
````

## 4. Inferência (Predição)

Executa o modelo em uma imagem e salva o resultado com detecções desenhadas.

````python
results = model('/home/lnknick/Downloads/1.jpg') #CAMINHO DA IMAGEM OU ARQUIVO PARA ANALISE
for i, r in enumerate(results):
    im_bgr = r.plot()
    im_rgb = Image.fromarray(im_bgr[..., ::-1])
    r.show()
    r.save(filename=f'results{i}.jpg')
````
